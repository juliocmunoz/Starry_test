Definimos una cuerpo muy parecido al de la tierra.

## Definimos el sistema

Ahora se pueden definir los parámetros del sistema. El elemento primario es una estrella que se comportará como el sol, rotando sobre su eje con un periodo de rotación de 24.47 días. El segundo elemento es un planeta que cuenta con las características de la tierra, pero tiene una órbita más cercana a su estrella (elemento primario), con un periodo orbital de 1 mes.

``````python
A = dict(
    ydeg=5,  
    udeg=2,  
    inc=80.0,  # inclinación
    amp=1.0,  # amplitudu (ñuminosidad)
    r=1.0, #Radio solar
    m=1.0,  # Masa solar
    prot= 24.47 ,  # período rotacional en días
    u=[0.40, 0.25],  # 
    y=A_y,  # 
)

B = dict(
    ydeg=20,  # 
    udeg=2,  # 
    inc=80.0,  # inclinación
    r=0.009,  #  radio solar
    m= 0.003,  #  masa solar
    porb= 30.0,  # período orbital (días)
    prot=1.0,  # períodorotacional (días)
    t0= 0.0,  # tiempo de referencia, cuando transita A
    u=[0.20, 0.05],  #
    y=B_y,  # 
)
``````

## Creación del Sistema

Una vez teniendo los atributos para cada uno de los planetas que conforarán el sistema, comenzamos a determinar el elemento primario y el secundario.

``````python
pri = starry.Primary(
    starry.Map(ydeg=A["ydeg"], udeg=A["udeg"], inc=A["inc"], amp=A["amp"]),
    r=A["r"],
    m=A["m"],
    prot=A["prot"],
)
pri.map[1:] = A["u"]
pri.map[1:, :] = A["y"]
pri.map.show(theta=np.linspace(0, 360, 60))
``````

``````python
sec = starry.Secondary(
    starry.Map(ydeg=B["ydeg"], udeg=B["udeg"], inc=B["inc"]),
    r=B["r"],
    m=B["m"],
    porb=B["porb"],
    prot=B["prot"],
    t0=B["t0"],
    inc=B["inc"],
)
sec.map.load("earth", sigma=0.08)
sec.map[1:] = B["u"]
sec.map.show(theta=np.linspace(0, 360, 50))
``````

Se genera un elemento **System** con los atributos que se declararon previante y se especifica los elementos que conformarán este sistema.

``````python
sys = starry.System(pri, sec)
``````

Se puede generar la simulación de como los elementos que conforman el sistema interactuan en el mismo. Es decir, podemos obserar la órbita y la rotación de cada uno de los elementos.

``````python
sys.show(t=np.linspace(0, 90, 700), window_pad=30, figsize=(10, 10))
``````

De esta forma, podemos generar las curvas de luz para analizarlas. Especificando el rango de días y el número de observaciones realizadas.

``````python
t = np.linspace(85, 90, 10000)
flux_true = sys.flux(t)
sigma = 0.0005
flux = flux_true + sigma * np.random.randn(len(t))
``````

``````python
fig, ax = plt.subplots(1, figsize=(12, 5))
ax.plot(t, flux, "k.", alpha=0.3, ms=2)
ax.plot(t, flux_true, lw=1)
ax.set_xlabel("tiempo [días]", fontsize=24)
ax.set_ylabel("flux", fontsize=24);
``````

``````python
t = np.linspace(0, 90, 10000)
flux_true = sys.flux(t)
sigma = 0.0005
flux = flux_true + sigma * np.random.randn(len(t))
``````

``````python
fig, ax = plt.subplots(1, figsize=(12, 5))
ax.plot(t, flux, "k.", alpha=0.3, ms=2)
ax.plot(t, flux_true, lw=1)
ax.set_xlabel("tiempo [días]", fontsize=24)
ax.set_ylabel("flux", fontsize=24);
``````

Además, se pueden obtener las curvas individuales de cada uno de los elementos que conforman el sistema definido.

``````python
flux_star, flux_planet = sys.flux(t, total=False)
``````

``````python
fig, ax = plt.subplots(1, figsize=(12, 5))
plt.plot(t, flux_star)
plt.xlabel("tiempo [días]", fontsize=24)
plt.ylabel("Estrella flux", fontsize=24);
``````

``````python
fig, ax = plt.subplots(1, figsize=(12, 5))
plt.plot(t, flux_planet)
plt.xlabel("tiempo [días]", fontsize=24)
plt.ylabel("Planeta flux", fontsize=24);
``````

La información de las curvas de luz se almacena en un objeto **Numpy** que puede ser guardado en sistema o se puede tener acceso a cada elemento que conforma este objeto.Es decir, se puede acceder a los arreglos y obtener su información.

``````python
np.savez("eb.npz", A=A, B=B, t=t, flux=flux, sigma=sigma)
``````
